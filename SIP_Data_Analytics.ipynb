**Approach Combination Of Trees And Depth**

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
n_estimators_list=[100,200,300,400,500,600,700]
max_depth_list=[2,3,4,5,6,7,8]
min_samples_leaf=[1,2,4,6,8]
results=[]
for n_trees in n_estimators_list:
  for depth in max_depth_list:
    for leaf in min_samples_leaf:
        rf=RandomForestClassifier(n_estimators=n_trees, max_depth=depth, min_samples_leaf=leaf, random_state=42, class_weight='balanced')
        rf.fit(X_train, Y_train)
        y_pred=rf.predict(X_test)
        accuracy=accuracy_score(Y_test, y_pred)
        recall=recall_score(Y_test, y_pred)
        precision=precision_score(Y_test, y_pred)
        f1=f1_score(Y_test, y_pred)
        results.append({'Trees':n_trees, 'Depth':depth, 'Leaf':leaf, 'Accuracy':accuracy, 'Recall':recall, 'Precision':precision, 'F1 Score':f1})
result_df=pd.DataFrame(results)
result_df=result_df.sort_values(by=['Recall', 'Accuracy'], ascending=False)
print(result_df.head(20))
mean_recall = result_df['Recall'].mean()
print(mean_recall)
```

**Customer Segmentation**

```python
segmentation_data=pd.DataFrame({'Tenure Months':X['Tenure Months'],
                                'Monthly Charges':X['Monthly Charges'],
                                'Total Charges':X['Total Charges'],
                                'Churn Probability':churn_probability})
```

```python
segmentation_data
```

**Implementation Of KMeans**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaled_data=scaler.fit_transform(segmentation_data)
print(scaled_data[:5])
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range (1,16):
  kmeans=KMeans(n_clusters=k, random_state=42)
  kmeans.fit(scaled_data)
  wcss.append(kmeans.inertia_)
  plt.figure(figsize=(8,6))
plt.plot(range(1,16), wcss, marker='o')
plt.xlabel('Number of Clusters K')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3, random_state=42)
```

```python
clusters=kmeans.fit_predict(scaled_data)
```

```python
segmentation_data
```

```python
segmentation_data['Cluster']=clusters
```

```python
segmentation_data
```

```python
cluster_summary=segmentation_data.groupby('Cluster').mean()
```

```python
cluster_summary
```

```python
cluster_names={
    0:'Budget Loyal Customers',
    1:'High Risk New Customers',
    2:'Loyal Premium Customers'
}
```

```python
segmentation_data['Cluster Segment']=segmentation_data['Cluster'].map(cluster_names)
```
